<h1>Runtime Visualization</h1>
<h2>Importing Libraries</h2>

```python
import pandas as pd
import matplotlib.pyplot as plt
```

<h2>Loading Runtime Data into Pandas DataFrame</h2>

```python
data = pd.DataFrame([10, 100, 1000, 10000, 100000], columns=['Iterations'])
data['Sequential'] = [1.834193828, 2.528033116, 19.913700793, 192.660934320, 1932.976192911]
data['2 CPUs'] = [5.453805815, 2.053744424, 13.033333829, 123.774489564, 1226.976072172]
data['4 CPUs'] = [0.854413334, 1.384999406, 6.744654116, 60.431221408, 595.324040617]
data['8 CPUs'] = [0.855997299, 1.134397737, 3.863426839, 31.097520716, 302.955319222]
data['16 CPUs'] = [1.032866468, 1.118882446, 2.584270231, 16.734743479, 157.457703631]

data.set_index('Iterations')
```

<h2>Runtime Plot</h2>

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 5))

for col_label in data.columns:
    if col_label == 'Iterations':
        continue
    ax.plot(data['Iterations'][:], data[col_label][:], label=col_label)
    ax.scatter(data['Iterations'][:], data[col_label][:])

ax.set_title("Parallel vs. Sequential Runtime")
ax.set_xlabel("Number of Iterations")
ax.set_ylabel("Runtime in Seconds")

fig.legend()

plt.savefig('Runtime-Plot.png')

fig.tight_layout()
```

<h2>Runtime Table</h2>

```python
fig, ax = plt.subplots(1,1, figsize=(10, 5))

fig.patch.set_visible(False)
ax.axis('off')
ax.axis('tight')

ax.table(cellText=data.values, colLabels=data.columns, loc='center')

plt.savefig('Runtime-Table.png')

fig.tight_layout()
```
